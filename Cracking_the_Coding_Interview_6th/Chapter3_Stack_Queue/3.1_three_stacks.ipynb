Describe how you could use a single array to implement three stacks.

```python
class Three_Stacks:
    def __init__(self,n=3):
        self.items = []
        for i in range(n):
            self.items.append(list())
    def isEmpty(self,stack_no):
        return self.items[stack_no] == []
    def push(self, data, stack_no ):
        self.items[stack_no].insert(0,data)
    def pop(self,stack_no):
        return self.items[stack_no].pop(0)
    def peek(self,stack_no):
        return self.items[stack_no][0]
    def size(self,stack_no):
        return len(self.items[stack_no])
    def print(self,stack_no):
        print(self.items[stack_no])
```

```python
three_stacks = Three_Stacks()
three_stacks.push(1,0)
three_stacks.push(2,0)
three_stacks.push(3,0)
three_stacks.push('a',1)
three_stacks.push('b',1)
three_stacks.push('c',1)
three_stacks.push(31,2)
three_stacks.push(32,2)
three_stacks.push(33,2)
```

```python
three_stacks.print(2)
```
